# Time Line of number of people kiled in Nort America because of Terrorism 

```python
# create a dataset a.
a=df[df['region']=='North America']
a.head()
```

```python
# create a datset a_a.
a_a = (a.groupby(['country','year'])['killed'].size()).reset_index()
a_a
```

```python
# LinePlot which compare the number of people killed in North American Countries.
fig = plt.figure(figsize=(16,5))
sns.lineplot(data = a_a,x = 'year',y = 'killed',hue = 'country',style = 'country',palette='magma_r',markers=True, dashes=False)
plt.title("Total People Kiled across the Years in Nort America",fontsize=20)
plt.xlabel('Name of the Year',fontsize=15)
plt.ylabel('Number of People Killed',fontsize=15)
plt.grid(True)
plt.show()
```


It can be observed from the provided information that during the 1970s, terrorist activities resulted in the highest recorded fatalities in North America. However, a positive trend is evident in the subsequent years, indicating a decline in the loss of life due to terrorist attacks in the region. This decline could potentially be attributed to the implementation of enhanced security measures across North America.
